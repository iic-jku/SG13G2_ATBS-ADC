```python
# Draw Programmable Capacitor
import matplotlib
matplotlib.rcParams.update({
    "text.usetex": True,
    "font.family": "lmodern"
})
import schemdraw as sd
from schemdraw import logic
import schemdraw.elements as elm
sd.svgconfig.svg2 = False
with sd.Drawing() as d:
    d.config(unit=2)
    d.config(fontsize=14)
    
    # Va
    elm.Dot(open=True).label('$V_\mathrm{a}$', loc='left', ofst=0.15)
    elm.Line().right().length(1).dot()
    d.push()
    
    # X3
    elm.Line().up().length(3.5)
    elm.Line().right().length(1.5)
    X3 = logic.Tgate().label('$X_3$', ofst=(0.7, 0.4)).anchor('out').theta(180)
    elm.Line().at(X3.c).up().length(0.25)
    elm.Dot(open=True).label(r'$V_\mathrm{EN3}$', loc='top', ofst=0.15)
    elm.Line().at(X3.cbar).down().length(0.25)
    elm.Dot(open=True).label(r'$\overline{V_\mathrm{EN3}}$', loc='bottom', ofst=0.15)

    # C3
    elm.Line().at(X3.in1).right().length(1)
    elm.Capacitor().label('$C_3$', loc='top', ofst=0.15).label(r'165\,pF', loc='bottom', ofst=0.15)
    elm.Line().right().length(1)
    elm.Line().down().length(3.5)

    d.pop()
    d.push()

    # X2
    elm.Line().right().length(1.5)
    X2 = logic.Tgate().label('$X_2$', ofst=(0.7, 0.4)).anchor('out').theta(180)
    elm.Line().at(X2.c).up().length(0.25)
    elm.Dot(open=True).label(r'$V_\mathrm{EN2}$', loc='top', ofst=0.15)
    elm.Line().at(X2.cbar).down().length(0.25)
    elm.Dot(open=True).label(r'$\overline{V_\mathrm{EN2}}$', loc='bottom', ofst=0.15)

    # C2
    elm.Line().at(X2.in1).right().length(1)
    elm.Capacitor().label('$C_2$', loc='top', ofst=0.15).label(r'132\,pF', loc='bottom', ofst=0.15)
    elm.Line().right().length(1)

    d.pop()
    d.push()

    # X1
    elm.Line().down().length(3.5)
    elm.Line().right().length(1.5)
    X1 = logic.Tgate().label('$X_1$', ofst=(0.7, 0.4)).anchor('out').theta(180)
    elm.Line().at(X1.c).up().length(0.25)
    elm.Dot(open=True).label(r'$V_\mathrm{EN1}$', loc='top', ofst=0.15)
    elm.Line().at(X1.cbar).down().length(0.25)
    elm.Dot(open=True).label(r'$\overline{V_\mathrm{EN1}}$', loc='bottom', ofst=0.15)

    # C1
    elm.Line().at(X1.in1).right().length(1)
    elm.Capacitor().label('$C_1$', loc='top', ofst=0.15).label(r'33\,pF', loc='bottom', ofst=0.15)
    elm.Line().right().length(1)
    elm.Line().up().length(3.5).dot()

    # Vb
    elm.Line().right().length(1)
    elm.Dot(open=True).label('$V_\mathrm{b}$', loc='right', ofst=0.15)

    # Save the schematic
    d.save('figures/programmable_cap/programmable_cap_circuit.svg')
    d.save('figures/programmable_cap/programmable_cap_circuit.pdf')
```
