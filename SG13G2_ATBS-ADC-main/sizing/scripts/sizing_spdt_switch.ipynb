# SPDT Switch without High-Z

```python
# Draw SPDT Switch without High-Z without W/L
import matplotlib
matplotlib.rcParams.update({
    "text.usetex": True,
    "font.family": "lmodern"
})
import schemdraw as sd
from schemdraw import logic
import schemdraw.elements as elm
sd.svgconfig.svg2 = False
with sd.Drawing() as d:
    d.config(unit=2)
    d.config(fontsize=14)
    
    # Va
    elm.Dot(open=True).label('$V_\mathrm{a}$', loc='left', ofst=0.15)
    elm.Line().right().length(1)
    
    # Left TG
    X1 = logic.Tgate().label('$X_1$', ofst=(0.7, 0.4)).anchor('out').theta(180)
    elm.Line().at(X1.c).up().length(0.5)
    elm.Dot(open=True).label(r'$V_\mathrm{CTRL}$', loc='top', ofst=0.15)
    elm.Line().at(X1.cbar).down().length(0.5)
    elm.Dot(open=True).label(r'$\overline{V_\mathrm{CTRL}}$', loc='bottom', ofst=0.15)

    # Middle
    elm.Line().at(X1.in1).right().length(1).dot()
    d.push()
    elm.Line().right().length(1)

    # Right TG
    X2 = logic.Tgate().label('$X_2$', ofst=(0.7, 0.4)).anchor('out').theta(180)
    elm.Line().at(X2.c).up().length(0.5)
    elm.Dot(open=True).label(r'$\overline{V_\mathrm{CTRL}}$', loc='top', ofst=0.15)
    elm.Line().at(X2.cbar).down().length(0.5)
    elm.Dot(open=True).label(r'$V_\mathrm{CTRL}$', loc='bottom', ofst=0.15)

    # Vb
    elm.Line().at(X2.in1).right().length(1)
    elm.Dot(open=True).label('$V_\mathrm{b}$', loc='right', ofst=0.15)

    # Vc
    d.pop()
    elm.Line().up().toy(X2.c)
    elm.Line().up().length(0.5)
    elm.Dot(open=True).label('$V_\mathrm{c}$', loc='top', ofst=0.15)

    # Save the schematic
    d.save('figures/spdt_switch/spdt_switch_circuit.svg')
    d.save('figures/spdt_switch/spdt_switch_circuit.pdf')
```

# SPDT Switch with High-Z

```python
# Draw SPDT Switch with High-Z without W/L
import matplotlib
matplotlib.rcParams.update({
    "text.usetex": True,
    "font.family": "lmodern"
})
import schemdraw as sd
from schemdraw import logic
import schemdraw.elements as elm
sd.svgconfig.svg2 = False
with sd.Drawing() as d:
    d.config(unit=2)
    d.config(fontsize=14)
    
    # Va
    elm.Dot(open=True).label('$V_\mathrm{a}$', loc='left', ofst=0.15)
    elm.Line().right().length(1)
    
    # Left TG
    X1 = logic.Tgate().label('$X_1$', ofst=(0.7, 0.4)).anchor('out').theta(180)
    elm.Line().at(X1.c).up().length(0.5)
    elm.Dot(open=True).label(r'$V_\mathrm{CTRL}$', loc='top', ofst=0.15)
    elm.Line().at(X1.cbar).down().length(0.5)
    elm.Dot(open=True).label(r'$\overline{V_\mathrm{CTRL}}$', loc='bottom', ofst=0.15)

    # Middle
    elm.Line().at(X1.in1).right().length(1).dot()
    d.push()
    elm.Line().right().length(1)

    # Right TG
    X2 = logic.Tgate().label('$X_2$', ofst=(0.7, 0.4)).anchor('out').theta(180)
    elm.Line().at(X2.c).up().length(0.5)
    elm.Dot(open=True).label(r'$\overline{V_\mathrm{CTRL}}$', loc='top', ofst=0.15)
    elm.Line().at(X2.cbar).down().length(0.5)
    elm.Dot(open=True).label(r'$V_\mathrm{CTRL}$', loc='bottom', ofst=0.15)

    # Vb
    elm.Line().at(X2.in1).right().length(1)
    elm.Dot(open=True).label('$V_\mathrm{b}$', loc='right', ofst=0.15)

    # Middle TG
    d.pop()
    elm.Line().up().toy(X2.c)
    elm.Line().up().length(1.5)
    X3 = logic.Tgate().label('$X_3$', ofst=(-0.65, 0.4)).anchor('out').theta(270)
    elm.Line().at(X3.c).left().length(0.5)
    elm.Dot(open=True).label(r'$\overline{V_\mathrm{HZ}}$', loc='left', ofst=0.15)
    elm.Line().at(X3.cbar).right().length(0.5)
    elm.Dot(open=True).label(r'$V_\mathrm{HZ}$', loc='right', ofst=0.15)

    # Vc
    elm.Line().at(X3.in1).up().length(1)
    elm.Dot(open=True).label('$V_\mathrm{c}$', loc='top', ofst=0.15)

    # Save the schematic
    d.save('figures/spdt_switch/spdt_hz_switch_circuit.svg')
    d.save('figures/spdt_switch/spdt_hz_switch_circuit.pdf')
```
