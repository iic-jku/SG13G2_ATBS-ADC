```python
# Draw Switched Capacitor Feedback Resistor
import matplotlib
matplotlib.rcParams.update({
    "text.usetex": True,
    "font.family": "lmodern"
})
import schemdraw as sd
from schemdraw import logic
import schemdraw.elements as elm
sd.svgconfig.svg2 = False
with sd.Drawing() as d:
    d.config(unit=2)
    d.config(fontsize=14)
    
    # Va
    elm.Dot(open=True).label('$V_\mathrm{a}$', loc='left', ofst=0.15)
    elm.Line().right().length(1)
    
    # X1
    X1 = logic.Tgate().label('$X_1$', ofst=(0.7, 0.4)).anchor('out').theta(180)
    elm.Line().at(X1.c).up().length(0.5)
    elm.Dot(open=True).label(r'$\Phi_1$', loc='top', ofst=0.15)
    elm.Line().at(X1.cbar).down().length(0.5)
    elm.Dot(open=True).label(r'$\overline{\Phi_1}$', loc='bottom', ofst=0.15)

    # C1
    elm.Line().at(X1.in1).right().length(1.25).dot()
    d.push()
    elm.Line().down().length(0.5)
    elm.Capacitor().label('$C_\mathrm{s}$', loc='top', ofst=0.15).label(r'4\,fF', loc='bottom', ofst=0.15)
    elm.GroundSignal()
    d.pop()
    elm.Line().right().length(1.25)

    # X2
    X2 = logic.Tgate().label('$X_2$', ofst=(0.7, 0.4)).anchor('out').theta(180)
    elm.Line().at(X2.c).up().length(0.5)
    elm.Dot(open=True).label(r'$\Phi_2$', loc='top', ofst=0.15)
    elm.Line().at(X2.cbar).down().length(0.5)
    elm.Dot(open=True).label(r'$\overline{\Phi_2}$', loc='bottom', ofst=0.15)

    # C2
    elm.Line().at(X2.in1).right().length(1.25).dot()
    d.push()
    elm.Line().down().length(0.5)
    elm.Capacitor().label('$C_\mathrm{s}$', loc='top', ofst=0.15).label(r'4\,fF', loc='bottom', ofst=0.15)
    elm.GroundSignal()
    d.pop()
    elm.Line().right().length(1.25)

    # X3
    X3 = logic.Tgate().label('$X_3$', ofst=(0.7, 0.4)).anchor('out').theta(180)
    elm.Line().at(X3.c).up().length(0.5)
    elm.Dot(open=True).label(r'$\Phi_1$', loc='top', ofst=0.15)
    elm.Line().at(X3.cbar).down().length(0.5)
    elm.Dot(open=True).label(r'$\overline{\Phi_1}$', loc='bottom', ofst=0.15)

    # C3
    elm.Line().at(X3.in1).right().length(1.25).dot()
    d.push()
    elm.Line().down().length(0.5)
    elm.Capacitor().label('$C_\mathrm{s}$', loc='top', ofst=0.15).label(r'4\,fF', loc='bottom', ofst=0.15)
    elm.GroundSignal()
    d.pop()
    elm.Line().right().length(1.25)

    # X4
    X4 = logic.Tgate().label('$X_4$', ofst=(0.7, 0.4)).anchor('out').theta(180)
    elm.Line().at(X4.c).up().length(0.5)
    elm.Dot(open=True).label(r'$\Phi_2$', loc='top', ofst=0.15)
    elm.Line().at(X4.cbar).down().length(0.5)
    elm.Dot(open=True).label(r'$\overline{\Phi_2}$', loc='bottom', ofst=0.15)

    # C4
    elm.Line().at(X4.in1).right().length(1.25).dot()
    d.push()
    elm.Line().down().length(0.5)
    elm.Capacitor().label('$C_\mathrm{s}$', loc='top', ofst=0.15).label(r'4\,fF', loc='bottom', ofst=0.15)
    elm.GroundSignal()
    d.pop()
    elm.Line().right().length(1.25)

    # X5
    X5 = logic.Tgate().label('$X_5$', ofst=(0.7, 0.4)).anchor('out').theta(180)
    elm.Line().at(X5.c).up().length(0.5)
    elm.Dot(open=True).label(r'$\Phi_1$', loc='top', ofst=0.15)
    elm.Line().at(X5.cbar).down().length(0.5)
    elm.Dot(open=True).label(r'$\overline{\Phi_1}$', loc='bottom', ofst=0.15)

    # C5
    elm.Line().at(X5.in1).right().length(1.25).dot()
    d.push()
    elm.Line().down().length(0.5)
    elm.Capacitor().label('$C_\mathrm{s}$', loc='top', ofst=0.15).label(r'4\,fF', loc='bottom', ofst=0.15)
    elm.GroundSignal()
    d.pop()
    elm.Line().right().length(1.25)

    # X6
    X6 = logic.Tgate().label('$X_6$', ofst=(0.7, 0.4)).anchor('out').theta(180)
    elm.Line().at(X6.c).up().length(0.5)
    elm.Dot(open=True).label(r'$\Phi_2$', loc='top', ofst=0.15)
    elm.Line().at(X6.cbar).down().length(0.5)
    elm.Dot(open=True).label(r'$\overline{\Phi_2}$', loc='bottom', ofst=0.15)

    # Vb
    elm.Line().at(X6.in1).right().length(1)
    elm.Dot(open=True).label('$V_\mathrm{b}$', loc='right', ofst=0.15)

    # Save the schematic
    d.save('figures/sc_fb_resistor/sc_fb_resistor_circuit.svg')
    d.save('figures/sc_fb_resistor/sc_fb_resistor_circuit.pdf')
```

```python
# Packages
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as sc
```

```python
# Variables
fclock = 8e6
# Two different fphi can be set via digital core
fphi = fclock / 2**10 # 7.81kHz
# fphi = fclock / 2**11 # 3.91kHz
C_shunt = 4e-15 # 4fF like the unit C of C1 of the CR-DAC
C_series = C_shunt
C_2 = 3.3e-12 # see sizing_biosignal_amplifier.ipynb
N = 5 # amount of stages

print('fclock =', round(float(fclock * 1e-6), 2), 'MHz')
print('fphi =', round(float(fphi * 1e-3), 6), 'kHz')
print('C_shunt =', round(float(C_shunt * 1e15), 2), 'fF')
print('C_series =', round(float(C_series * 1e15), 2), 'fF')
print('C_2 =', round(float(C_2 * 1e12), 2), 'pF')
```

```python
# Calculation of the equivalent resistor for the shunt-type topology
R_el_shunt = N / (fphi * C_shunt)
print('R_el_shunt =', round(float(R_el_shunt * 1e-9), 2), 'GOhm')
```

```python
# Calculation of the equivalent resistor for the series-type topology
R_el_series = (N - 1) / (fphi * C_series)
print('R_el_series =', round(float(R_el_series * 1e-9), 2), 'GOhm')
```

```python
# Calculation of the cut-off frequency with shunt-type topology and C_2
f_c = 1 / (2 * np.pi * R_el_shunt * C_2)
print('f_c =', round(float(f_c), 2), 'Hz')
```
