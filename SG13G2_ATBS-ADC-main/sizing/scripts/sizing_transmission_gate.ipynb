```python
# Draw Transmission Gate without dummy transistors and without W/L
import matplotlib
matplotlib.rcParams.update({
    "text.usetex": True,
    "font.family": "lmodern"
})
import schemdraw as sd
import schemdraw.elements as elm
sd.svgconfig.svg2 = False
with sd.Drawing() as d:
    d.config(unit=2)
    d.config(fontsize=14)
    
    # Input
    elm.Dot(open=True).label('$V_\mathrm{a}$', loc='left', ofst=0.15)
    elm.Line().right().length(1).dot()
    d.push()

    # PMOS
    elm.Line().up().length(1)
    elm.Line().right().length(1)
    M2 = elm.AnalogPFet(offset_gate=False).anchor('drain').theta(270).label('$M_2$', ofst=0.25).reverse()
    elm.Line().at(M2.gate).up().length(0.5)
    elm.Dot(open=True).label(r'$\overline{V_\mathrm{CTRL}}$', loc='top', ofst=0.15)
    elm.Line().at(M2.source).right().length(1)
    elm.Line().down().length(1).dot()

    # Output
    elm.Line().right().length(1)
    elm.Dot(open=True).label('$V_\mathrm{b}$', loc='right', ofst=0.15)

    # NMOS
    d.pop()
    elm.Line().down().length(1)
    elm.Line().right().length(1)
    M1 = elm.AnalogNFet(offset_gate=False).anchor('drain').theta(90).label('$M_1$', ofst=-1.4).reverse()
    elm.Line().at(M1.gate).down().length(0.5)
    elm.Dot(open=True).label(r'$V_\mathrm{CTRL}$', loc='bottom', ofst=0.15)
    elm.Line().at(M1.source).right().length(1)
    elm.Line().up().length(1)

    # Save the schematic
    d.save('figures/transmission_gate/TG_circuit_wo_dummy.svg')
    d.save('figures/transmission_gate/TG_circuit_wo_dummy.pdf')
```

```python
# Draw Transmission Gate with dummy transistors without W/L
import matplotlib
matplotlib.rcParams.update({
    "text.usetex": True,
    "font.family": "lmodern"
})
import schemdraw as sd
import schemdraw.elements as elm
sd.svgconfig.svg2 = False
with sd.Drawing() as d:
    d.config(unit=2)
    d.config(fontsize=14)
    
    # Input
    elm.Dot(open=True).label('$V_\mathrm{a}$', loc='left', ofst=0.15)
    elm.Line().right().length(1).dot()
    d.push()

    # PMOS
    elm.Line().up().length(1)
    elm.Line().right().length(1)
    M2 = elm.AnalogPFet(offset_gate=False).anchor('drain').theta(270).label('$M_2$', ofst=0.25).reverse()
    elm.Line().at(M2.gate).up().length(0.5)
    elm.Dot(open=True).label(r'$\overline{V_\mathrm{CTRL}}$', loc='top', ofst=0.15)
    elm.Line().at(M2.source).right().length(1)
    M2_dummy = elm.AnalogPFet(offset_gate=False).anchor('drain').theta(270).label('$M_{2D}$', ofst=0.40).reverse()
    elm.Line().at(M2_dummy.gate).up().length(0.5)
    elm.Dot(open=True).label(r'$V_\mathrm{CTRL}$', loc='top', ofst=0.15)
    elm.Dot().at(M2_dummy.drain)
    elm.Line().down().length(0.25)
    elm.Line().right().length(1.67)
    elm.Line().up().length(0.25)
    elm.Dot().at(M2_dummy.source)
    elm.Line().at(M2_dummy.source).right().length(1)
    elm.Line().down().length(1).dot()

    # Output
    elm.Line().right().length(1)
    elm.Dot(open=True).label('$V_\mathrm{b}$', loc='right', ofst=0.15)

    # NMOS
    d.pop()
    elm.Line().down().length(1)
    elm.Line().right().length(1)
    M1 = elm.AnalogNFet(offset_gate=False).anchor('drain').theta(90).label('$M_1$', ofst=-1.4).reverse()
    elm.Line().at(M1.gate).down().length(0.5)
    elm.Dot(open=True).label(r'$V_\mathrm{CTRL}$', loc='bottom', ofst=0.15)
    elm.Line().at(M1.source).right().length(1)
    M1_dummy = elm.AnalogNFet(offset_gate=False).anchor('drain').theta(90).label('$M_{1D}$', ofst=-1.7).reverse()
    elm.Line().at(M1_dummy.gate).down().length(0.5)
    elm.Dot(open=True).label(r'$\overline{V_\mathrm{CTRL}}$', loc='bottom', ofst=0.15)
    elm.Dot().at(M1_dummy.drain)
    elm.Line().up().length(0.25)
    elm.Line().right().length(1.67)
    elm.Line().down().length(0.25)
    elm.Dot().at(M1_dummy.source)
    elm.Line().at(M1_dummy.source).right().length(1)
    elm.Line().up().length(1)

    # Save the schematic
    d.save('figures/transmission_gate/TG_circuit_w_dummy.svg')
    d.save('figures/transmission_gate/TG_circuit_w_dummy.pdf')
```
