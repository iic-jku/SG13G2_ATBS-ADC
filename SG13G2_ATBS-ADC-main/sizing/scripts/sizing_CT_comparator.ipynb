## Ideal CT Comparator

```python
# Draw Ideal CT Comparator
import matplotlib
matplotlib.rcParams.update({
    "text.usetex": True,
    "font.family": "lmodern"
})
import schemdraw as sd
import schemdraw.elements as elm
import schemdraw.dsp as dsp
sd.svgconfig.svg2 = False
with sd.Drawing() as d:
    d.config(unit=2)
    d.config(fontsize=16)
    
    elm.Dot(open = True).label('$V_\mathrm{in,p}$', loc='left')
    elm.Line().right().length(2)
    d.push()
    elm.Line().down().length(1)
    elm.Resistor().label('$R_\mathrm{in}$')
    elm.Line().down().length(1)
    elm.Line().left().length(2)
    elm.Dot(open = True).label('$V_\mathrm{in,n}$', loc='left')

    d.pop()
    elm.Gap().right().length(2)
    d.push()
    elm.Line().down().length(1)
    elm.SourceV().down().label('ASRC', ofst=-2.3).reverse()
    elm.Line().down().length(1)
    elm.Line().right().length(6.5)
    elm.Dot(open = True).label('$V_\mathrm{SS}$', loc='right')

    d.pop()
    elm.Line().right().length(1.5)
    dsp.Square().label('$t_\mathrm{pd}$')
    elm.Line().right().length(1)
    elm.Resistor().label('$R_\mathrm{out}$')
    elm.Line().right().length(1)
    elm.Dot(open = True).label('$V_\mathrm{out}$', loc='right')

    # Save the schematic
    d.save('figures/comparator/continuous_time/CT_comparator_ideal.svg')
    d.save('figures/comparator/continuous_time/CT_comparator_ideal.pdf')
```
