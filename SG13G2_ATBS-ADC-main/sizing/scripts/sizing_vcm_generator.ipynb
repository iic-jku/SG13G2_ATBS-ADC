```python
# Draw Vcm Generator without W/L
import matplotlib
matplotlib.rcParams.update({
    "text.usetex": True,
    "font.family": "lmodern"
})
import schemdraw as sd
from schemdraw import logic
import schemdraw.elements as elm
sd.svgconfig.svg2 = False
with sd.Drawing() as d:
    d.config(unit=2)
    d.config(fontsize=14)
    
    # X1
    elm.Vdd().label('$V_\mathrm{DD}$', loc='top', ofst=0.15)
    elm.Line().down().length(0.5)
    X1 = logic.Tgate().label('$X_1$', ofst=(-0.65, 0.4)).anchor('in1')
    elm.Line().at(X1.c).left().length(0.5)
    elm.Dot(open=True).label(r'$\Phi_{10}$', loc='left', ofst=0.15)
    elm.Line().at(X1.cbar).right().length(0.5)
    elm.Dot(open=True).label(r'$\overline{\Phi_{10}}$', loc='right', ofst=0.15)
    elm.Line().at(X1.out).down().length(1).dot()

    # X2
    elm.Line().right().length(2.5)
    X2 = logic.Tgate().label('$X_2$', ofst=(0.7, 0.4)).anchor('out').theta(180)
    elm.Line().at(X2.c).up().length(0.5)
    elm.Dot(open=True).label(r'$\Phi_9$', loc='top', ofst=0.15)
    elm.Line().at(X2.cbar).down().length(0.5)
    elm.Dot(open=True).label(r'$\overline{\Phi_9}$', loc='bottom', ofst=0.15)
    elm.Line().at(X2.in1).right().length(1.0).dot()
    d.push()
    elm.Line().right().length(1)
    elm.Dot(open=True).label(r'$V_\mathrm{cm}$', loc='right', ofst=0.15)

    # X4
    d.pop()
    elm.Line().down().length(5.5)
    elm.Line().left().length(1.0)
    X4 = logic.Tgate().label('$X_4$', ofst=(0.7, 0.4)).anchor('in1')
    elm.Line().at(X4.c).up().length(0.5)
    elm.Dot(open=True).label(r'$\Phi_9$', loc='top', ofst=0.15)
    elm.Line().at(X4.cbar).down().length(0.5)
    elm.Dot(open=True).label(r'$\overline{\Phi_9}$', loc='bottom', ofst=0.15)
    elm.Line().at(X4.out).left().length(2.5).dot()
    d.push()

    # C2
    elm.Line().down().length(0.5)
    elm.Capacitor().label('$C_2$')
    GNDpoint = d.here
    elm.GroundSignal()
    
    # X3
    d.pop()
    elm.Line().up().length(1)
    X3 = logic.Tgate().label('$X_3$', ofst=(-0.65, 0.4)).anchor('out').theta(270)
    elm.Line().at(X3.c).left().length(0.5)
    elm.Dot(open=True).label(r'$\Phi_{10}$', loc='left', ofst=0.15)
    elm.Line().at(X3.cbar).right().length(0.5)
    elm.Dot(open=True).label(r'$\overline{\Phi_{10}}$', loc='right', ofst=0.15)
    elm.Line().at(X3.in1).up().length(1).dot()
    d.push()

    # C1
    elm.Line().up().length(0.5)
    elm.Capacitor().label('$C_1$')
    elm.Line().up().length(0.5)
    
    # X5
    d.pop()
    elm.Line().left().length(3)
    elm.Line().down().length(2.3)
    X5 = logic.Tgate().label('$X_5$', ofst=(-0.65, 0.4)).anchor('in1')
    elm.Line().at(X5.c).left().length(0.5)
    elm.Dot(open=True).label(r'$\Phi_9$', loc='left', ofst=0.15)
    elm.Line().at(X5.cbar).right().length(0.5)
    elm.Dot(open=True).label(r'$\overline{\Phi_9}$', loc='right', ofst=0.15)
    elm.Line().at(X5.out).down().toy(GNDpoint)
    elm.GroundSignal()

    # Save the schematic
    d.save('figures/vcm_generator/vcm_generator_circuit.svg')
    d.save('figures/vcm_generator/vcm_generator_circuit.pdf')
```
