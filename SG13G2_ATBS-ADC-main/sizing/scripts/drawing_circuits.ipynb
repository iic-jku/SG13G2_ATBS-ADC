```python
# Draw block diagram of a standard AFE
import matplotlib
matplotlib.rcParams.update({
    "text.usetex": True,
    "font.family": "lmodern"
})
import schemdraw as sd
import schemdraw.elements as elm
import schemdraw.dsp as dsp
sd.svgconfig.svg2 = False
with sd.Drawing() as d:
    d.config(unit=2)
    d.config(fontsize=16)
    
    elm.Dot(open = True).label('$V_\mathrm{in}$', loc='left')
    elm.Line().right().length(1)
    dsp.Amp().label('LNA', loc='bottom', ofst=0.25)
    elm.Line().right().length(1)
    dsp.Filter(response='lp').label('AAF', loc='bottom', ofst=0.25)
    elm.Line().right().length(1)
    dsp.Adc().label('ADC', loc='bottom', ofst=0.25)
    elm.Line().right().length(1)
    dsp.Square().label('FPGA', loc='bottom', ofst=0.25)
    elm.Line().right().length(1)
    elm.Dot(open = True).label('$01011010$', loc='right')

    # Save the schematic
    d.save('figures/analog_front_end/AFE_block_diagram.svg')
    d.save('figures/analog_front_end/AFE_block_diagram.pdf')
```

```python
# Draw Return Ratio Framework
import matplotlib
matplotlib.rcParams.update({
    "text.usetex": True,
    "font.family": "lmodern"
})
import schemdraw as sd
import schemdraw.elements as elm
import schemdraw.dsp as dsp
sd.svgconfig.svg2 = False
with sd.Drawing() as d:
    d.config(unit=2)
    d.config(fontsize=16)
    
    # Input + SumSigma
    elm.Dot(open=True).label('$X(j \omega)$')
    elm.Line().right().length(d.unit*0.75).dot()
    dot1 = d.here
    elm.Arrow().right().length(d.unit*0.75)
    sm1 = dsp.SumSigma()

    # H(jw) + SumSigma
    dsp.Arrow().right()
    dsp.Box().label(r'$H_\mathrm{ol}(j \omega)$')
    elm.Arrow().right()
    sm2 = dsp.SumSigma()
    elm.Line().right().dot()
    dot2 = d.here
    d.push() 
    elm.Line().right()
    elm.Dot(open=True).label('$Y(j \omega)$')

    # b
    d.pop()  
    elm.Line().down().length(d.unit*1.5)
    dsp.Arrow().left().length(d.unit*1.90)
    dsp.Box().label(r'$\beta$')
    elm.Line().tox(sm1.S)
    elm.Line().up().length(d.unit*0.75)
    dsp.Arrow().toy(sm1.S).label('$-$', loc='bottom')

    # d
    elm.Line().at(dot1).up().length(d.unit*1.5)
    dsp.Arrow().right().length(d.unit*1.90)
    dsp.Box().label(r'$d$')
    elm.Line().tox(sm2.N)
    elm.Arrow().toy(sm2.N)

    # Save the schematic
    d.save('figures/appendix/Return_Ratio_Framework.svg')
    d.save('figures/appendix/Return_Ratio_Framework.pdf')
```

```python
# Draw Biosignal Amplifier (singled-ended) without RC values
# This circuit should illustrate the derivation of the total and simplified transfer function
import matplotlib
matplotlib.rcParams.update({
    "text.usetex": True,
    "font.family": "lmodern"
})
import schemdraw as sd
import schemdraw.elements as elm
sd.svgconfig.svg2 = False
with sd.Drawing() as d:
    d.config(unit=2)
    d.config(fontsize=16)
    
    elm.GroundSignal(lead=False)
    GND = d.here
    elm.Line().up().length(0.5)
    elm.SourceSin().label('$V_\mathrm{in}$')
    elm.Line().up().length(0.5)
    elm.Line().right().length(0.5)
    elm.Resistor().label('$R_1$')
    elm.Capacitor().label('$C_1$')
    elm.Line().right().length(0.25).dot()
    dot1 = d.here
    d.push()

    elm.Line().right().length(0.25)
    OPA = elm.Opamp(leads=True).anchor('in1').theta(0)
    elm.Line().at(OPA.in2).left().length(0.25)
    elm.Line().down().toy(GND)
    elm.GroundSignal(lead=False)

    elm.Line().at(OPA.out).right().length(0.25).dot()
    dot2 = d.here
    elm.Line().right().length(1).dot(open=True)
    elm.Line().down().length(0.25).color('white')
    elm.Arrow().down().length(1.4).label('$V_\mathrm{out}$', ofst=-1)
    elm.Line().down().length(0.25).color('white')
    elm.Dot(open=True)
    elm.Line().down().toy(GND)
    elm.GroundSignal(lead=False)

    d.pop()
    elm.Line().up().length(1.5).dot()
    d.push()
    elm.Line().right().length(1)
    elm.Capacitor().label('$C_2$')
    elm.Line().right().tox(dot2).dot()

    d.pop()
    elm.Line().up().length(1.5)
    elm.Line().right().length(1)
    elm.Resistor().label('$R_2$')
    elm.Line().right().tox(dot2)
    elm.Line().down().toy(dot2)

    # Save the schematic
    d.save('figures/appendix/biosignal_amplifier_SE_circuit.svg')
    d.save('figures/appendix/biosignal_amplifier_SE_circuit.pdf')
```
